```python
import scipy
import scipy.signal
import matplotlib.pyplot as plt
import numpy
```

```python
def plotstep(G, name = ""):
    _, ys_step = G.step(T=ts)
    plt.plot(ts, ys_step, label = name);
```

```python
ts = numpy.linspace(0,20,1001)
```

```python
tau_a_span = [8 , 2, 1, -1, -4]
tau = 4

plt.figure(1,figsize=(10,10))
plt.figure(2,figsize=(10,10))

tau_a = tau_a_span[0]

num = ( tau_a, 1)
den = ( tau, 1)
func = scipy.signal.lti(num, den)

plt.figure(2)
poles = func.poles
plt.scatter( poles.real, poles.imag, marker = "d", color = "black", label = "Poles", s =100)

for tau_a in tau_a_span:
    plt.figure(1)
    num = ( tau_a, 1)
    den = ( tau, 1)
    func = scipy.signal.lti(num, den)
    plotstep(func, "A = " + str(tau_a))
    plt.legend(loc = "best")
    
    plt.figure(2)
    zeroes = func.zeros
    plt.scatter( zeroes.real, zeroes.imag, label = "Zeros for A = " + str(tau_a))
    plt.legend( loc = "best")
```
