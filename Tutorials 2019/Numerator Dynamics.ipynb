$$Figure 5.2 (using LTI)$$

```python
import scipy.signal
```

```python
ts2 = numpy.linspace(0, 20, 2000)
```

```python
G=[]
for a in τa:
    numerator = [K*M*a, K*M]
    denominator = [τ1, 1]
    G0 = scipy.signal.lti(numerator, denominator)
    G.append(G0)
```

```python
def plotstep(G):
    _, ys_step = G.step(T=ts)
    plt.plot(ts, ys_step);
```

```python
plt.figure(3)
plt.figure(figsize=(15,7))
plt.subplot(1,2,1)
plt.title('Figure 5.2')
for i in G:
    plotstep(i)
plt.subplot(1,2,2)
plt.title('Poles plot')
plt.scatter(-1/τ1, 0, marker='x', s=100,label='-1/τ1')
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.axis([-2.5, 2.5, -2.5, 2.5])
plt.ylabel('Imaginary Part')
plt.xlabel('Real Part')
plt.plot()    
```
