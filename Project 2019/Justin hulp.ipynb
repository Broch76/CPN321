```python
import tclab
import matplotlib as plt
%matplotlib inline
import numpy
```

```python
TCLab = tclab.setup(connected=False, speedup=100)
tfinal = 300
```

```python
def kool(t):
    return 10*numpy.sin(t/200) + 50
```

```python
def kool2(t, s):
    if t < s:
        return 0
    else:
        return 50
```

```python
with TCLab() as lab:
    h = tclab.Historian(lab.sources)
    p = tclab.Plotter(h, tfinal)
    for t in tclab.clock(tfinal, 1):
        lab.Q1(kool2(t, 100))
        p.update(t)
```
